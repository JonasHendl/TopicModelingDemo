```python
from datasets import load_dataset
from bertopic import BERTopic
import pandas as pd

dataset = load_dataset("app_reviews")
```

```python
dataset.set_format(type='pandas')
df = dataset['train'][:]
```

```python
telegram_reviews = df[df.package_name == 'org.telegram.messenger'].reset_index()
telegram_reviews.head()
```

```python

docs = telegram_reviews['review']
topic_model = BERTopic()
topics, probs = topic_model.fit_transform(docs)
```

```python
topic_model.visualize_topics()
```

```python
topic_model.visualize_documents(docs, sample=0.001)
```
